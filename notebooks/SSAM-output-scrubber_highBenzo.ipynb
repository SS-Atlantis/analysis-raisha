## High benzo, high resolution run starting at 1993

```python
control1993_file = "/ocean/rlovindeer/MOAD/analysis-raisha/SSmodel_outputs/Scrubber_discharge_3y_highres_control/outputSalishSea.nc"
scenario1993_1e6_file = "/ocean/rlovindeer/MOAD/analysis-raisha/SSmodel_outputs/Scrubber_discharge_3y_highres_benzo_1e6/outputSalishSea.nc"
scenario1993_1e4_file = "/ocean/rlovindeer/MOAD/analysis-raisha/SSmodel_outputs/Scrubber_discharge_3y_highres_benzo/outputSalishSea.nc"

control = xr.open_dataset(str(control1993_file), decode_cf=True)
scenario1993_1e6 = xr.open_dataset(str(scenario1993_1e6_file), decode_cf=True)
scenario1993_1e4 = xr.open_dataset(str(scenario1993_1e4_file), decode_cf=True)
t = np.ma.filled(scenario1993_1e6.variables['t'], np.nan)
```

```python
start = 360
end = 700
#end = t.size-1

event_start = np.datetime64('1993-01-20')
time = t[start:end]
```

```python
Benzo = np.ma.filled(scenario1993_1e6.variables['Benzo'], np.nan)
Benzo = Benzo.sum(axis=2)
Benzo = Benzo.sum(axis=1)

fig, ax = plt.subplots(figsize = (14,3))
ax.plot(time, Benzo[start:end], 'k')
ax.plot(event_start, 0, 'ro', alpha=0.5)
ax.set_ylabel('mg benzo m$^{-3}$');
```
